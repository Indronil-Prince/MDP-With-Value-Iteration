## MDP with Value Iteration

**States**

'S' : Starting State

'G' : Goal State

'X' : Danger State

'-' : Other state


**Policies**

'>' : Move East

'^' : Move North

's' : Stay

```python
import numpy as np # linear algebra
```

```python
def value_iteration(grid, p):
    
    cost = -1

    reward_goal = 10
    reward_danger = -10
    reward_other = -1

    row, col = np.shape(grid)[0], np.shape(grid)[1]
    v = np.zeros((row, col))
    pol = [['s' for _ in range(col)] for _ in range(row)]

#Value Initializaion
    for r in reversed(range(row)):
        for c in range(col):
            if grid[r][c]=='G':
                v[r][c]=reward_goal
                pol[r][c]='G'
            elif grid[r][c]=='X':
                v[r][c]=reward_danger
                pol[r][c]='X'
            elif grid[r][c]=='B':
                v[r][c]=0
                pol[r][c]='B'
            else:
                v[r][c]=reward_other
                pol[r][c]='s'

    print(v,"\n")
    
#Start Iteration
    for i in range(10000):
        vnew = np.zeros((row, col))
        for r in reversed(range(row)):
            for c in range(col):
                v0 = v[r][c]
                if grid[r][c] not in ['G','X','B']:
                    ver = v[r][c+1] if c<col-1 else v0 #Action east, outcome right
                    veu = v[r-1][c] if r>0 else v0 #Action east, outcome up
                    
                    vnl = v[r][c-1] if c>0 else v0 #Action north, outcome left
                    vnu = v[r-1][c] if r>0 else v0 #Action north, outcome up
                    
                    ve = (p*(cost+ver) + (1-p)*(cost+veu))
                    vn = (p*(cost+vnu) + (1-p)*(cost+vnl))
                    
                    vnew[r][c] = max(ve,vn)
                    if c==col-1:
                        pol[r][c] = '^' if vn>=ve else 's'
                    elif r==0:
                        pol[r][c] = '>' if ve>=vn else 's'
                    else:
                        pol[r][c] = '^' if vn>=ve else '>'
                else:
                    vnew[r][c]=v0
            
        print(vnew,"\n")
    
        equal = 0
        for r in reversed(range(row)):
            for c in range(col):
                if round(v[r][c],4)!=round(vnew[r][c],4):
                    equal = 1
        if equal == 0:
            print ("Done!!! ", i+1, " iterations\n\nPolicy:")
            for r2 in (range(row)):
                print(pol[r2])
            break
        else:
            v = vnew
```

```python
grid = [
    ['-','-','G'],
    ['X','-','-'],
    ['-','-','-'],
    ['S','-','-'],
]

p=0.9
value_iteration(grid, p)
```

```python
p=0.8
value_iteration(grid, p)
```

```python
p=0.99
value_iteration(grid, p)
```
